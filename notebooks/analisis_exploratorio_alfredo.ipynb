```python
# Carga de librerías

import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
```

```python
# Carga de datos
df = pd.read_csv("../data/raw/SocialMediaUsage_SleepLatencyAnalysis_Singapore.csv")
df
```

```python
# Redondeo de columnas flotantes a 3 digitos. 

for column in df.columns:
    if df[column].dtype == 'float64':
        df[column] = np.round(df[column], 3)
        
df.sample(5)
```

```python
# Cambio de nombres de las columnas a 'snake_case'

df.columns = [column.lower().replace(' ','_') for column in df.columns]
df.columns
```

```python
# Valores únicos de las variables categóricoas

for column in df.columns:
    if df[column].dtype == 'object':
        print('La columna {}, tiene los siguientes valores:'.format(column))
        print(df[column].value_counts().sort_values(ascending=False))
        plt.figure(figsize=(8, 6))
        sns.countplot(data=df, x=column, hue = column, legend = False, palette="viridis")
        plt.title(f"Distribución de valores en '{column}'")
        plt.xlabel(column)
        plt.ylabel("Frecuencia")
        print()
```

```python
# Lista de columnas numéricas

numeric_columns = []

for column in df.columns:
    if df[column].dtype == 'int64' or df[column].dtype == 'float64':
        numeric_columns.append(column)
```

```python
# Filtrado de columnas numéricas 

df_numerical = df[numeric_columns]
df_numerical = df_numerical.drop(columns='participant_id')
```

```python
for column in df_numerical.columns:
    print(f'Descripción de los valores de la columna {column}')
    print(df[column].describe())
    plt.figure(figsize=(8, 6))
    sns.violinplot(data=df_numerical, x=column, color ='cornflowerblue')
    plt.title(f"Distribución de la columna '{column}'")
    plt.xlabel(column)
    plt.ylabel("Valor")
    plt.show()
    print()
```

```python
# Verificación de correlación entre las variables numéricas 

correlation_matrix = df_numerical.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Mapa de calor de correlación")
plt.show()
```
