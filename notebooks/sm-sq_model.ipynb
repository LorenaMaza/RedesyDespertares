```python
import pandas as pd

# Cargar el dataset
smsq_df = pd.read_csv("..\\data\\raw\\SocialMediaUsage_SleepLatencyAnalysis_Singapore.csv")

# Mostar el dataset
smsq_df
```

```python
# Cambio de nombres de las columnas a 'snake_case'

smsq_df.columns = [column.lower().replace(' ','_') for column in smsq_df.columns]
smsq_df.columns
```

```python
# Borrar la columna "participant_id"
smsq_df.drop(columns=['participant_id'], inplace=True)
smsq_df.head()
```

```python
# Crear la columna "sleep_quality" categórica que represente la columna "sleep_quality_sleep_quality_rating"

# Definir la funcion para determinar las categorías en "sleep_quality"
def sq_categories(sleep_quality_rating):
    if sleep_quality_rating < 2:
        return 'very_bad'
    elif sleep_quality_rating < 3:
        return 'bad'
    elif sleep_quality_rating < 4:
        return 'medium'
    elif sleep_quality_rating < 5:
        return 'good'
    else:
        return 'very_good'

smsq_df['sleep_quality'] = smsq_df['sleep_quality_rating'].apply(sq_categories)

smsq_df.head()
```
